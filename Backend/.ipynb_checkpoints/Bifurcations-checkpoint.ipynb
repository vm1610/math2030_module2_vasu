## Bifurcation Diagram

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python


r_values = np.linspace(2.5, 4.0, 8000)   # fine resolution
n_iter = 1000                            # total iterations
last = 100                               # number of iterations to plot (after transient)
x0 = 0.5                                 # initial value

```

```python
# Data Collection 
x = x0 * np.ones(len(r_values))
results_r = []
results_x = []

for i in range(n_iter):
    x = r_values * x * (1 - x)
    if i >= (n_iter - last):            # skip transients
        results_r.extend(r_values)
        results_x.extend(x)


```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(results_r, results_x, ',k', alpha=0.2)  
plt.title("Bifurcation Diagram of the Logistic Map")
plt.xlabel("Growth rate $r$")
plt.ylabel("Population $x$")
plt.xlim(2.5, 4.0)
plt.tight_layout()
plt.savefig("Data/bifurcation_placeholder.png", dpi=160)
plt.show()
```

# Sensitivity to Initial Conditions

```python
# Parameters
r = 4.0
x0 = 0.5000
x1 = 0.5001
n = 50

# Allocate arrays
x_vals = np.empty(n)
y_vals = np.empty(n)
diff = np.empty(n)

# Initialize
x_vals[0] = x0
y_vals[0] = x1
diff[0] = abs(x0 - x1)
```

```python
# Iterate through logistic map
for i in range(1, n):
    x_vals[i] = r * x_vals[i-1] * (1 - x_vals[i-1])
    y_vals[i] = r * y_vals[i-1] * (1 - y_vals[i-1])
    diff[i] = abs(x_vals[i] - y_vals[i])
```

```python
# Plotting
plt.figure(figsize=(10, 5))

# Trajectories
plt.subplot(1, 2, 1)
plt.plot(range(n), x_vals, label='$x_0 = 0.5000$', lw=2)
plt.plot(range(n), y_vals, label='$x_0 = 0.5001$', lw=2, linestyle='--')
plt.title("Chaotic Trajectories ($r = 4.0$)")
plt.xlabel("Iteration $n$")
plt.ylabel("$x_n$")
plt.legend()

# Difference (log scale)
plt.subplot(1, 2, 2)
plt.semilogy(range(n), diff, color='red')
plt.title("Divergence: $|x_n - x'_n|$")
plt.xlabel("Iteration $n$")
plt.ylabel("Absolute Difference (log scale)")

plt.tight_layout()
plt.savefig("sensitivity_placeholder.png", dpi=300)
plt.show()
```

```python
# Feigenbaum automation: find r_n for periods 2,4,8,16,32; compute delta_n; save plots & CSVs
import numpy as np
import matplotlib
matplotlib.use('Agg')  # Use non-interactive backend
import matplotlib.pyplot as plt
import pandas as pd
import os

os.makedirs("BackendData", exist_ok=True)

# ---------- Core logistic iteration ----------
def logistic_iter(r, x0=0.5, n_transient=2500, n_keep=2048):
    x = x0
    total = n_transient + n_keep
    out = np.empty(n_keep)
    for i in range(total):
        x = r * x * (1 - x)
        if i >= n_transient:
            out[i - n_transient] = x
    return out

# ---------- Period estimator (powers of 2) ----------
def estimate_period_power2(xs, tol=1e-8, max_pow=10):
    """
    Rough period estimator that snaps to the nearest power-of-two cycle size.
    Works well for period-doubling cascades (2,4,8,16,...).
    """
    xr = np.round(xs, 12)  # merge machine noise
    uniq = []
    for v in np.sort(xr):
        if not uniq or abs(v - uniq[-1]) > tol:
            uniq.append(v)
    m = max(1, len(uniq))
    # snap m to nearest power of 2 up to 2^max_pow
    powers = [2**k for k in range(max_pow+1)]
    nearest = min(powers, key=lambda p: abs(p-m))
    return nearest

# ---------- Find first r where period >= target (coarse + bisection refine) ----------
def first_r_with_period(target_period, r_min=2.8, r_max=4.0, coarse=5000, x0=0.5):
    rs = np.linspace(r_min, r_max, coarse)
    hit = None
    # coarse sweep
    for r in rs:
        xs = logistic_iter(r, x0=x0, n_transient=2000, n_keep=1024)
        if estimate_period_power2(xs, max_pow=12) >= target_period:
            hit = r
            break
    if hit is None:
        return None
    # refine by bisection in a small window
    a = max(r_min, hit - 0.05)
    b = min(r_max, hit + 0.05)
    for _ in range(28):
        mid = 0.5*(a+b)
        p_mid = estimate_period_power2(logistic_iter(mid, n_transient=3000, n_keep=2048), max_pow=12)
        if p_mid >= target_period:
            b = mid
        else:
            a = mid
    return 0.5*(a+b)

# ---------- Compute r_n for periods 2,4,8,16,32 ----------
print("Computing bifurcation points...")
periods = [2,4,8,16,32]
r_points = []
for p in periods:
    print(f"Finding r for period {p}...")
    r_n = first_r_with_period(p)
    r_points.append(r_n)
    print(f"  r_{periods.index(p)+1} = {r_n}")

df_r = pd.DataFrame({"n":[1,2,3,4,5], "period":periods, "r_n":r_points})
df_r.to_csv("BackendData/feigenbaum_r_points.csv", index=False)
print("\nBifurcation Points:")
print(df_r)

# ---------- Compute Feigenbaum deltas ----------
print("\nComputing Feigenbaum deltas...")
deltas = []
for i in range(2, len(r_points)):
    num = r_points[i-1] - r_points[i-2]
    den = r_points[i]   - r_points[i-1]
    delta = num/den
    deltas.append(delta)
    print(f"  delta_{i} = {delta}")

df_delta = pd.DataFrame({
    "n":[2,3,4][:len(deltas)],
    "r_{n-2}": r_points[:-2],
    "r_{n-1}": r_points[1:-1],
    "r_n":     r_points[2:],
    "delta_n": deltas
})
df_delta.to_csv("BackendData/feigenbaum_deltas.csv", index=False)
print("\nFeigenbaum Deltas:")
print(df_delta)

# ---------- Plot: delta_n convergence ----------
print("\nGenerating delta convergence plot...")
plt.figure(figsize=(6,4))
plt.plot([2,3,4][:len(deltas)], deltas, marker='o', linestyle='-', linewidth=2, markersize=8)
plt.axhline(4.6692016091, linestyle='--', linewidth=1, color='red', label=r"Theoretical $\delta \approx 4.669$")
plt.xlabel(r"$n$  (for $\delta_n$)", fontsize=12)
plt.ylabel(r"$\delta_n$", fontsize=12)
plt.title("Convergence of Feigenbaum Ratios", fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("BackendData/feigenbaum_deltas_plot.png", dpi=300)
print("  Saved: BackendData/feigenbaum_deltas_plot.png")

# ---------- Optional: plot r_n points on a zoomed bifurcation background ----------
print("\nGenerating bifurcation diagram with r_n points...")
rmin, rmax = 2.9, 3.60
r_vals = np.linspace(rmin, rmax, 6500)
x = 0.5*np.ones_like(r_vals)
n_iter, last = 1400, 200
R, X = [], []
for i in range(n_iter):
    x = r_vals * x * (1 - x)
    if i >= (n_iter-last):
        R.append(r_vals.copy())
        X.append(x.copy())
R = np.concatenate(R)
X = np.concatenate(X)

plt.figure(figsize=(7.5,4))
plt.plot(R, X, ',k', alpha=0.25)
# Overlay the detected r_n inside the zoom window
for k, rn in enumerate(r_points, start=1):
    if rn is not None and (rmin <= rn <= rmax):
        plt.axvline(rn, color='crimson', linestyle='--', linewidth=1)
        plt.text(rn+0.002, 0.92, f"$r_{k}$", rotation=90, color='crimson', fontsize=9)
plt.xlim(rmin, rmax)
plt.ylim(0,1)
plt.title("Zoomed Bifurcation with Detected $r_n$ Points", fontsize=14)
plt.xlabel("$r$", fontsize=12)
plt.ylabel("$x$", fontsize=12)
plt.tight_layout()
plt.savefig("BackendData/feigenbaum_zoom_points.png", dpi=300)
print("  Saved: BackendData/feigenbaum_zoom_points.png")

print("\n" + "="*60)
print("All files saved successfully!")
print("="*60)
print("Files created:")
print("  - BackendData/feigenbaum_r_points.csv")
print("  - BackendData/feigenbaum_deltas.csv")
print("  - BackendData/feigenbaum_deltas_plot.png")
print("  - BackendData/feigenbaum_zoom_points.png")
```
